```python
import pandas as pd
from functools import reduce

pd.set_option('display.max_rows', None)

dates = pd.read_csv('csv_from_scrape/baseball_dates_index.csv')
```

```python
fisher_2017 = pd.read_csv('csv_from_scrape/fisher_2017_df.csv')
fisher_2017.columns.values[1] = 'DATE'
fisher_2017.columns.values[2] = '2017 Batting Averages'
fisher_2017.drop('Unnamed: 0', axis = 1)
```

```python
frames = [dates, fisher_2017]
drops = ['Unnamed: 0']
```

```python
df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['DATE'], how='outer'), frames)
```

```python
fisher_batting = df_merged.drop(drops, axis = 1)
```

```python
fisher_batting.to_csv('data_aggregated/fisher_batting.csv', index=False)
```
