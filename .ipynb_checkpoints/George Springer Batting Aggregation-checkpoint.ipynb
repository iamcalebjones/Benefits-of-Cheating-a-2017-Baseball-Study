```python
import pandas as pd
from functools import reduce

pd.set_option('display.max_rows', None)

dates = pd.read_csv('csv_from_scrape/baseball_dates_index.csv')
```

```python
springer_2014 = pd.read_csv('csv_from_scrape/springer_2014_df.csv')
springer_2014.columns.values[1] = 'DATE'
springer_2014.columns.values[2] = '2014 Batting Averages'
springer_2014.drop('Unnamed: 0', axis = 1)
```

```python
springer_2015 = pd.read_csv('csv_from_scrape/springer_2015_df.csv')
springer_2015.columns.values[1] = 'DATE'
springer_2015.columns.values[2] = '2015 Batting Averages'
springer_2015.drop('Unnamed: 0', axis = 1)
```

```python
springer_2016 = pd.read_csv('csv_from_scrape/springer_2016_df.csv')
springer_2016.columns.values[1] = 'DATE'
springer_2016.columns.values[2] = '2016 Batting Averages'
springer_2016.drop('Unnamed: 0', axis = 1)
```

```python
springer_2017 = pd.read_csv('csv_from_scrape/springer_2017_df.csv')
springer_2017.columns.values[1] = 'DATE'
springer_2017.columns.values[2] = '2017 Batting Averages'
springer_2017.drop('Unnamed: 0', axis = 1)
```

```python
frames = [dates, springer_2014, springer_2015, springer_2016, springer_2017]
drops = ['Unnamed: 0_x', 'Unnamed: 0_y']
```

```python
df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['DATE'], how='outer'), frames)
```

```python
springer_batting = df_merged.drop(drops, axis = 1)
```

```python
springer_batting.to_csv('data_aggregated/springer_batting.csv', index = False)
```
