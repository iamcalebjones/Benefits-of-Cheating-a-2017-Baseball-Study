```python
import pandas as pd
from functools import reduce

pd.set_option('display.max_rows', None)

dates = pd.read_csv('csv_from_scrape/baseball_dates_index.csv')
```

```python
davis_2017 = pd.read_csv('csv_from_scrape/davis_2017_df.csv')
davis_2017.columns.values[1] = 'DATE'
davis_2017.columns.values[2] = '2017 Batting Averages'
davis_2017.drop('Unnamed: 0', axis = 1)
```

```python
frames = [dates, davis_2017]
drops = ['Unnamed: 0']
```

```python
df_merged = reduce(lambda left,right: pd.merge(left,right,on=['DATE'], how='outer'), frames)
```

```python
davis_batting = df_merged.drop(drops, axis=1)
```

```python
davis_batting.to_csv('data_aggregated/davis_batting.csv', index=False)
```
