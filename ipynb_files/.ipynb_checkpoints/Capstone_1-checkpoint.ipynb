```python
better_dates = []
for i in range(len(bangs)):
    
    x = bangs['game_date'][i].split('/')
    
    if len(x[0]) == 1 and len(x[1]) == 1:
        better_dates.append('0' + x[1] + '-0' + x[0])
        
    elif len(x[0]) > 1 and len(x[1]) == 1:
        better_dates.append('0' + x[1] + '-' + x[0])
        
    elif len(x[0]) == 1 and len(x[1]) > 1:
        better_dates.append(x[1] + '-0' + x[0])
        
    else:
        better_dates.append(x[1] + '-' + x[0])
```

```python
better_dates
```

```python
bangs['DATE'] = pd.Series(better_dates)
```

Some ideas for next steps are to see what percentage of pitches had bangs for the entire data set, and then similarly for each batter.

```python
has_bangs = (bangs.has_bangs.values == 'y').sum()
total_pitches = len(bangs.has_bangs)
percent_bangs = (has_bangs/total_pitches)*100

print(f'"There were {has_bangs} pitches with bangs out of {total_pitches} total pitches analyzed in the dataset, totaling {percent_bangs:.2f}% with bangs."')
```
