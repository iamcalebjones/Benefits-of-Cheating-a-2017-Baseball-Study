```python
import pandas as pd
```

```python
mo = ['04', '05', '06', '07', '08', '09']
days = list(range(1,32))
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
baseball_year = []
for i in mo:
    for j in days:
        if len(str(j)) == 1:
            baseball_year.append(i + '-0' + str(j))
        else:
            baseball_year.append(i + '-' +str(j))
```

```python
baseball_year.pop(baseball_year.index('04-31'))
baseball_year.pop(baseball_year.index('06-31'))
baseball_year.pop(baseball_year.index('09-31'))
```

```python
baseball_year.append('10-01')
baseball_year.append('10-02')
baseball_year.append('10-03')
baseball_year.append('10-04')
baseball_year.append('10-05')
```

```python
print('april index is ' + str(baseball_year.index('04-01')))
print('may index is ' + str(baseball_year.index('05-01')))
print('june index is ' + str(baseball_year.index('06-01')))
print('july index is ' + str(baseball_year.index('07-01')))
print('august index is ' + str(baseball_year.index('08-01')))
print('september index is ' + str(baseball_year.index('09-01')))
print('october index is ' + str(baseball_year.index('10-01')))
```

```python
baseball_year.insert(0, '03-31')
```

```python
baseball_year.insert(0, '03-30')
```

```python
baseball_year
```

```python
baseball_months = ['nan']*len(baseball_year)
```

```python
baseball_months[2] = 'April'
```

```python
baseball_months[32] = 'May'
```

```python
baseball_months[63] = 'June'
baseball_months[93] = 'July'
baseball_months[124] = 'August'
baseball_months[155] = 'September'
baseball_months[185] = 'October'
```

```python
d = {'DATE': baseball_year, 'MONTHS': baseball_months}
```

```python
baseball_year_index = pd.DataFrame(data = d)
```

```python
baseball_year_index
```

```python
baseball_year_index.to_csv('../csv_from_scrape/baseball_dates_index.csv', index=False)
```
