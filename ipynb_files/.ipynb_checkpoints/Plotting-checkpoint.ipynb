## Plotting Layout

```python
# set up plotting framework before incorporating a loop

player_batting = pd.read_csv('../data_aggregated/altuve_batting.csv')
player_batting.fillna(method = 'ffill', inplace=True)
player_bangs = bangs[(bangs['batter'] == 'Jose Altuve') & (bangs['has_bangs'] == 'y')]

better_dates_bangs = pd.to_datetime(player_bangs['DATE'], format='%m-%d')
player_bangs.insert(loc=0, column='Better Date', value=better_dates_bangs)

better_dates = pd.to_datetime(player_batting['DATE'], format='%m-%d')
player_batting = player_batting.drop('DATE', axis = 1)
player_batting.insert(loc=0, column='DATE', value=better_dates)

# initialize the plot
fig, ax = plt.subplots(figsize = (20, 8))

# loop through columns and plot for each batting average column
for j in player_batting.columns[1:-1]:
    ax.plot(player_batting['DATE'], player_batting[j], label = j[0:4])

# exclude 2017 from this plot loop for specifying extra formatting
ax.plot(player_batting['DATE'], player_batting['2017 Batting Averages'], label = '2017', color='black',  linewidth = 2.5)

# vlines are for the pitches with bangs
ax.vlines(player_bangs['Better Date'], ymin = 0, ymax = 0.550, ls='--', label = 'Signs Stolen')

#fuck around with getting xticks to be the months only
locator = mdates.MonthLocator()
fmt = mdates.DateFormatter('%b')
X = plt.gca().xaxis
X.set_major_locator(locator)
X.set_major_formatter(fmt)

# format the axes
ax.yaxis.set_major_formatter(FormatStrFormatter('%.3f'))
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)

# extend x limits by date
d = timedelta(days = 5)
start_date = min(player_batting['DATE']) - d
end_date = max(player_batting['DATE']) + d

# set plot limits
plt.xlim(start_date, end_date)
plt.ylim(0, .6)

# set player variable for file saving stuff
player = 'Jose Altuve'

ax.legend(loc = 'lower right', fontsize = 12)
# plt.title(player + " Career Batting Averages", fontsize = 20)
plt.title(player + " Career Batting Averages with 2017 Stolen Signs", fontsize = 22)
# plt.title(player + " 2017 Batting Average with Stolen Signs", fontsize = 20)
plt.xlabel("MLB Season Dates", fontsize = 18, labelpad=10)
plt.ylabel('Batting Average', fontsize = 18, labelpad=10)

#     plt.savefig('../plots_and_images/' + i + '_2017_batting_averages_stolen_signs.png')
plt.show()
```

```python
# dictionary of players i want to plot and their file paths for importing their data

batters_to_plot = {'Alex Bregman': '../data_aggregated/bregman_batting.csv',
                   'Carlos Beltran': '../data_aggregated/Beltran_batting.csv',
                   'Carlos Correa': '../data_aggregated/correa_batting.csv',
                   'Evan Gattis': '../data_aggregated/gattis_batting.csv',
                   'George Springer': '../data_aggregated/springer_batting.csv',
                   'Jake Marisnick': '../data_aggregated/marisnick_batting.csv',
                   'Jose Altuve': '../data_aggregated/altuve_batting.csv',
                   'Marwin Gonzalez': '../data_aggregated/gonzales_batting.csv',
                   'Yuli Gurriel': '../data_aggregated/gurriel_batting.csv'}
```

```python
# test looping through dictionary

for i in batters_to_plot:
    print(i)
```

```python
# loop for generating plots for each batters career with stolen signs

for i in batters_to_plot:
    
    player_batting = pd.read_csv(batters_to_plot[i])
    player_batting.fillna(method = 'ffill', inplace=True)
    player_bangs = bangs[(bangs['batter'] == i) & (bangs['has_bangs'] == 'y')]
    
    better_dates_bangs = pd.to_datetime(player_bangs['DATE'], format='%m-%d')
    player_bangs.insert(loc=0, column='Better Date', value=better_dates_bangs)
    
    better_dates = pd.to_datetime(player_batting['DATE'], format='%m-%d')
    player_batting = player_batting.drop('DATE', axis = 1)
    player_batting.insert(loc=0, column='DATE', value=better_dates)
    
    # initialize the plot
    fig, ax = plt.subplots(figsize = (20, 8))
    
    # loop through columns and plot for each batting average column
    for j in player_batting.columns[1:-1]:
        ax.plot(player_batting['DATE'], player_batting[j], label = j[0:4])
    
    # exclude 2017 from the plot loop so i can specify extra formatting
    ax.plot(player_batting['DATE'], player_batting['2017 Batting Averages'], label = '2017', color='black',  linewidth = 2.5)

    ax.vlines(player_bangs['Better Date'], ymin = 0, ymax = 0.550, ls='--', label = 'Signs Stolen')

    #fuck around with getting xticks to be the months only
    locator = mdates.MonthLocator()
    fmt = mdates.DateFormatter('%b')
    X = plt.gca().xaxis
    X.set_major_locator(locator)
    X.set_major_formatter(fmt)

    # format the axes
    ax.yaxis.set_major_formatter(FormatStrFormatter('%.3f'))
    plt.xticks(fontsize = 14)
    plt.yticks(fontsize = 14)

    # extend x limits by date
    d = timedelta(days = 5)
    start_date = min(player_batting['DATE']) - d
    end_date = max(player_batting['DATE']) + d

    # set plot limits
    plt.xlim(start_date, end_date)
    plt.ylim(0, .6)

    # set player variable for file saving stuff
    player = i

    ax.legend(loc = 'lower right', fontsize = 12)
    plt.title(player + " Career Batting Averages with 2017 Stolen Signs", fontsize = 22)
    plt.xlabel("MLB Season Dates", fontsize = 18, labelpad=10)
    plt.ylabel('Batting Average', fontsize = 18, labelpad=10)

    plt.savefig('../plots_and_images/' + i + '_career_batting_averages_stolen_signs.png')
    plt.show()
```
