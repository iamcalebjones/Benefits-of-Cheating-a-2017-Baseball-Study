# Cameron Maybin

Played for the following teams before the Astros in 2017:

- 2008-2010: Florida Marlins
- 2011-2014: San Diego Padres
- 2015: Atlanta Braves

```python
urls = ['https://www.retrosheet.org/boxesetc/2007/Imaybc0010012007.htm',
        'https://www.retrosheet.org/boxesetc/2009/Imaybc0010032009.htm',
        'https://www.retrosheet.org/boxesetc/2010/Imaybc0010042010.htm',
        'https://www.retrosheet.org/boxesetc/2011/Imaybc0010052011.htm',
        'https://www.retrosheet.org/boxesetc/2012/Imaybc0010062012.htm',
        'https://www.retrosheet.org/boxesetc/2013/Imaybc0010072013.htm',
        'https://www.retrosheet.org/boxesetc/2014/Imaybc0010082014.htm',
        'https://www.retrosheet.org/boxesetc/2015/Imaybc0010092015.htm',
        'https://www.retrosheet.org/boxesetc/2016/Imaybc0010102016.htm',
        'https://www.retrosheet.org/boxesetc/2017/Imaybc0010112017.htm',
        'https://www.retrosheet.org/boxesetc/2017/Imaybc0010122017.htm'
       ]

csv_names = ['csv_from_scrape/maybin_2007_df.csv',
             'csv_from_scrape/maybin_2009_df.csv',
             'csv_from_scrape/maybin_2010_df.csv',
             'csv_from_scrape/maybin_2011_df.csv',
             'csv_from_scrape/maybin_2012_df.csv',
             'csv_from_scrape/maybin_2013_df.csv',
             'csv_from_scrape/maybin_2014_df.csv',
             'csv_from_scrape/maybin_2015_df.csv',
             'csv_from_scrape/maybin_2016_df.csv',
             'csv_from_scrape/maybin_2017_1_df.csv',
             'csv_from_scrape/maybin_2017_2_df.csv'
            ]

years = ['2007', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2017']
```

```python
len(urls) == len(csv_names) == len(years)
```

```python
for idx, url in enumerate(urls):
    
    # batting averages

    xpath = "//pre[5]/text()"
    r = requests.get(url)
    page = html.parse(StringIO(r.text))
    batter = page.xpath(xpath)
    
    batter_game_data_only = []
    for a, row in enumerate(batter):
        if a % 2 == 0:
            batter_game_data_only.append(row)
    
    batter_game_data_cleaned = []
    for row in batter_game_data_only[1:]:
        if len(row) > len(batter_game_data_only[2]):
            batter_game_data_cleaned.append(row[0:112])
        else:
            batter_game_data_cleaned.append(row)
        
    batter_averages = [float(row[87:91]) for row in batter_game_data_cleaned]

    # dates played this year

    soup = BeautifulSoup(r.text, 'html.parser')
    
    link_text = []
    for a in soup.find_all('a'):
        link_text.append(a.string)
        
    batter_dates = []
    for i in link_text:
        if years[idx] in i:
            batter_dates.append(i)
            
    batter_dates_cleaned = []
    for date in batter_dates:
        batter_dates_cleaned.append(date.replace(' ', '0'))
        
    batter_dates_final = []
    for date in batter_dates_cleaned:
        batter_dates_final.append(datetime.strftime(datetime.strptime(date, '%m-%d-%Y'), '%m-%d'))
        
    batter_df = pd.DataFrame(batter_averages, batter_dates_final).reset_index()
    batter_df.columns = [years[idx], 'Batting Average']
    batter_df.to_csv(csv_names[idx])
    batter_df.plot(years[idx], 'Batting Average')
```
