```python
# define categories for each of the plate appearance events
# a successful event is getting on base (home run included, side effect of home run is scoring lol) or driving in a run
# a failed event is the at bat ending in an out
# a non-contributing event is anyting that wouldn't contribute the to batting average/at-bat tally

ab_success = ['Home Run',
              'Single',
              'Field Error',
              'Double',
              'Triple',
              'Sac Fly',
              'Sac Bunt']

ab_failure = ['Strikeout',
              'Double Play',
              'Grounded Into DP',
              'Flyout',
              'Groundout',
              'Lineout',
              'Runner Out',
              'Forceout',
              'Pop Out',
              'Sac Bunt',
              'Strikeout - DP',
              'Fielders Choice Out',
              'Bunt Groundout',
              'Bunt Pop Out']

ab_non_contribute = ['Catcher Interference',
                     'Walk',
                     'Fan interference',
                     'Hit By Pitch',
                     'Batter Interference']
```

```python
# define categories for each pitch description
# a successful description is anything that is either the batter getting on base, which does not contribute to ending the at bat (balls)
# a failed description is antying that ends the at bat (in play out) or contributes to its end (strikes, foul balls)
# a non-contributing description is a pitch at that does not contribute to the batting average calculation

desc_success = ['In play, run(s)',
                'In play, no out',
                'Ball',
                'Ball In Dirt',
                'Pitchout']

desc_failure = ['Called Strike',
                'Swinging Strike',
                'In play, out(s)',
                'Swinging Strike (Blocked)',
                'Missed Bunt',
                'Foul',
                'Foul Bunt',
                'Foul Tip',
                'Foul (Runner Going)']

desc_non_contribute = ['Hit By Pitch']
```

```python
batters_to_evaluate = ['Alex Bregman',
                       'Carlos Beltran',
                       'Carlos Correa',
                       'Evan Gattis',
                       'George Springer',
                       'Jake Marisnick',
                       'Jose Altuve',
                       'Marwin Gonzalez',
                       'Yuli Gurriel']
```

```python
'''the AB metrics are a little more realistic of a metric, because they describe how the at bat ended, either in
getting on base and potentially scoring a run, or in getting put out.

The per pitch metrics are a little more misleading because lets say a batter like Alex Bregman fouls off 5 pitches
and then homers, there would be 5 failures and then one success in the successful AB. So i might drop the per pitch
metrics altogether.'''

evaluation_metrics = ['Pitches with Bangs',
                      'Total Pitches',
                      'Percent Bangs',
                      'Bang AB Success Rate',
                      'Clean AB Success Rate',
                      'Bang AB Failure Rate',
                      'Clean AB Failure Rate',
                      'Bang AB Runs Generated',
                      'Clean AB Runs Generated',
                      'Bang Pitch Success Rate',
                      'Clean Pitch Succes Rate',
                      'Bang Pitch Failure Rate',
                      'Clean Pitch Failure Rate',
                      'Bang Pitch Runs Generated',
                      'Clean Pitch Runs Generated']
```

```python
# and putting those into a loop for each player...

# batters_to_evaluate = ['Jose Altuve']

player_evaluation_df = pd.DataFrame()
player_evaluation_df['Metric'] = evaluation_metrics


for name in batters_to_evaluate:
    player_bangs = bangs[(bangs['batter'] == name) & (bangs['has_bangs'] == 'y')]
    player_no_bangs = bangs[(bangs['batter'] == name) & (bangs['has_bangs'] == 'n')]
    
    # initialize dictionaries for successes and failures of bangs and no bangs at bats and pitches
    bang_ab_dict = {'success': 0, 'failure': 0, 'na': 0}
    bang_pitch_dict = {'success': 0, 'failure': 0, 'na': 0}
    no_bang_ab_dict = {'success': 0, 'failure': 0, 'na': 0}
    no_bang_pitch_dict = {'success': 0, 'failure': 0, 'na': 0}
    
    # loop for at-bats with bangs
    for i in player_bangs['at_bat_event']:
        if i in ab_success:
            bang_ab_dict['success'] += 1
        elif i in ab_failure:
            bang_ab_dict['failure'] += 1
        else:
            bang_ab_dict['na'] += 1

    # loop for pitches with bangs
    for i in player_bangs['description']:
        if i in desc_success:
            bang_pitch_dict['success'] += 1
        elif i in desc_failure:
            bang_pitch_dict['failure'] += 1
        else:
            bang_pitch_dict['na'] += 1
            
    # loop for at-bats without bangs
    for i in player_no_bangs['at_bat_event']:
        if i in ab_success:
            no_bang_ab_dict['success'] += 1
        elif i in ab_failure:
            no_bang_ab_dict['failure'] += 1
        else:
            no_bang_ab_dict['na'] += 1
    
    # loop for pitches without bangs
    for i in player_no_bangs['description']:
        if i in desc_success:
            no_bang_pitch_dict['success'] += 1
        elif i in desc_failure:
            no_bang_pitch_dict['failure'] += 1
        else:
            no_bang_pitch_dict['na'] += 1
            
            
            
    # calculate metrics, populate a list, put list in pandas dataframe
    
    # bang_ab_success - an at-bat with bangs that was successful
    bang_ab_success = round((bang_ab_dict['success'] / sum(bang_ab_dict.values())) * 100, 2)
    
    # clean_ab_success - an at-bat without bangs that was successful
    clean_ab_success = round((no_bang_ab_dict['success'] / sum(no_bang_ab_dict.values())) * 100, 2)
    
    # bang_ab_failure - an at-bat with bangs that was an out
    bang_ab_failure = round((bang_ab_dict['failure'] / sum(bang_ab_dict.values())) * 100, 2)
    
    # clean_ab_failure - an at-bat without bangs that was an out
    clean_ab_failure = round((no_bang_ab_dict['failure'] / sum(no_bang_ab_dict.values())) * 100, 2)
    
    # bang_ab_runs_generated - how many at-bats with bangs resulted in runs being scored (these are effectively RBIs)
    # (homer, single + runs scored, double + runs etc...)
    # loop through some boolean indexing and populate an accumulator
    bang_ab_runs_generated = 0
    for i in ab_success:
        bang_ab_runs_generated += (len(player_bangs[(player_bangs['at_bat_event'] == i) & (player_bangs['description'] == 'In play, run(s)')]))
    
    # clean_ab_runs_generated - how many at-bats without bangs resulted in runs being scored (these are effectively RBIs)
    # (homer, single + runs scored, double + runs etc...)
    # same idea, loop and accumulate
    clean_ab_runs_generated = 0
    for i in ab_success:
        clean_ab_runs_generated += (len(player_no_bangs[(player_no_bangs['at_bat_event'] == i) & (player_no_bangs['description'] == 'In play, run(s)')]))
        
    # bang_pitch_success - successful pitches with bangs
    bang_pitch_success = round((bang_pitch_dict['success'] / sum(bang_pitch_dict.values())) * 100, 2)
    
    # clean_pitch_success - successful pitches without bangs
    clean_pitch_success = round((no_bang_pitch_dict['success'] / sum(no_bang_pitch_dict.values())) * 100, 2)
    
    # bang_pitch_failure - failure pitches with bangs
    bang_pitch_failure = round((bang_pitch_dict['failure'] / sum(bang_pitch_dict.values())) * 100, 2)
    
    # clean_pitch_failure - successful pitches without bangs
    clean_pitch_failure = round((no_bang_pitch_dict['failure'] / sum(no_bang_pitch_dict.values())) * 100, 2)
    
    # bang_pitch_runs_generated - how many pitches with bangs resulted in runs being scored (runs from fielding errors)
    bang_pitch_runs_generated = (len(player_bangs[player_bangs['description'] == 'In play, run(s)']))
    
    # clean_pitch_runs_generated - how many pitches without bangs resulted in runs being scored
    # (in play + run(s))
    # same idea, loop and accumulate
    clean_pitch_runs_generated = (len(player_no_bangs[player_no_bangs['description'] == 'In play, run(s)']))
        
    # put the calculated metrics into a list, put list into dataframe of players
    metric = [batters_bangs_dict[name], batters_pitches_dict[name], percent_bangs_by_batter_dict[name], bang_ab_success, clean_ab_success, bang_ab_failure, clean_ab_failure, bang_ab_runs_generated, 
              clean_ab_runs_generated, bang_pitch_success, clean_pitch_success, bang_pitch_failure, 
              clean_pitch_failure, int(bang_pitch_runs_generated), clean_pitch_runs_generated]
    
    player_evaluation_df[name] = metric
    
```
