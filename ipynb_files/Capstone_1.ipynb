```python
better_dates = []
for i in range(len(bangs)):
    
    x = bangs['game_date'][i].split('/')
    
    if len(x[0]) == 1 and len(x[1]) == 1:
        better_dates.append('0' + x[1] + '-0' + x[0])
        
    elif len(x[0]) > 1 and len(x[1]) == 1:
        better_dates.append('0' + x[1] + '-' + x[0])
        
    elif len(x[0]) == 1 and len(x[1]) > 1:
        better_dates.append(x[1] + '-0' + x[0])
        
    else:
        better_dates.append(x[1] + '-' + x[0])
```

```python
bangs['DATE'] = pd.Series(better_dates)
```

```python
bangs_better_date = pd.read_csv('../bangs_file_better_date.csv')
bangs_better_date.head() # this dataframe has a new column with better date formatting, last column on the far right
```

I'm going to create some lists of the unique values in columns of interest.

```python
batters = list(pd.unique(bangs_better_date['batter']))
batters.sort()
batters
```

```python
percent_bangs_by_batter = [] # populate a list with tuples of names and percent bangs so it can be sorted
percent_bangs_by_batter_dict = {} # also a dictionary for easy searching...
for i in batters:
    bang_pitches = ((bangs_better_date[(bangs_better_date.batter == i)]).has_bangs.values == 'y').sum()
    total_pitches = len((bangs_better_date[(bangs_better_date.batter == i)]))
    percent_bangs = (bang_pitches/total_pitches) * 100
    percent_bangs_by_batter.append((i, round(percent_bangs, 2)))
    percent_bangs_by_batter_dict[i] = round(percent_bangs, 2)

percent_bangs_by_batter = sorted(percent_bangs_by_batter, key=lambda x: x[1])
```

```python
batters_pitches = [] # populate a list with tuples of names and total number of pitches seen so it can be sorted
batters_pitches_dict = {} # also a dictionary for easy searching...
for i in batters:
    total_pitches = len((bangs_better_date[(bangs_better_date.batter == i)]))
    batters_pitches.append((i, total_pitches))
    batters_pitches_dict[i] = total_pitches
    
batters_pitches = sorted(batters_pitches, key=lambda x: x[1])
```

```python
batters_bangs = [] # populate a list with tuples of names and total number of pitches with bangs so it can be sorted
batters_bangs_dict = {} # also a dictionary for easy searching...
for i in batters:
    bang_pitches = ((bangs_better_date[(bangs_better_date.batter == i)]).has_bangs.values == 'y').sum()
    batters_bangs.append((i, bang_pitches))
    batters_bangs_dict[i] = bang_pitches

batters_bangs = sorted(batters_bangs, key=lambda x: x[1]) 
```

```python
# select batters who had higher than 15% bangs and more than 300 pitches

for i in batters:
    if percent_bangs_by_batter_dict[i] > 15 and batters_pitches_dict[i] > 300:
        print(i)
```
