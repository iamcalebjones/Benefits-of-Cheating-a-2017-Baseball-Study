```python
import pandas as pd
from functools import reduce
```

```python
pd.set_option('display.max_rows', None)
```

```python
dates = pd.read_csv('csv_from_scrape/baseball_dates_index.csv')
```

```python
beltran_2012 = pd.read_csv('csv_from_scrape/beltran_2012_df.csv')
beltran_2012.columns.values[1] = 'DATE'
beltran_2012.columns.values[2] = '2012 Batting Averages'
beltran_2012.drop('Unnamed: 0', axis = 1)
```

```python
beltran_2013 = pd.read_csv('csv_from_scrape/beltran_2013_df.csv')
beltran_2013.columns.values[1] = 'DATE'
beltran_2013.columns.values[2] = '2013 Batting Averages'
beltran_2013.drop('Unnamed: 0', axis = 1)
```

```python
beltran_2014 = pd.read_csv('csv_from_scrape/beltran_2014_df.csv')
beltran_2014.columns.values[1] = 'DATE'
beltran_2014.columns.values[2] = '2014 Batting Averages'
beltran_2014.drop('Unnamed: 0', axis = 1)
```

```python
beltran_2015 = pd.read_csv('csv_from_scrape/beltran_2015_df.csv')
beltran_2015.columns.values[1] = 'DATE'
beltran_2015.columns.values[2] = '2015 Batting Averages'
beltran_2015.drop('Unnamed: 0', axis = 1)
```

```python
beltran_2016_1 = pd.read_csv('csv_from_scrape/beltran_2016_1_df.csv')
beltran_2016_1.columns.values[1] = 'DATE'
beltran_2016_1.columns.values[2] = '2016 Batting Averages 1'
beltran_2016_1.drop('Unnamed: 0', axis = 1)
```

```python
beltran_2016_2 = pd.read_csv('csv_from_scrape/beltran_2016_2_df.csv')
beltran_2016_2.columns.values[1] = 'DATE'
beltran_2016_2.columns.values[2] = '2016 Batting Averages 2'
beltran_2016_2.drop('Unnamed: 0', axis = 1)
```

```python
beltran_2017 = pd.read_csv('csv_from_scrape/beltran_2017_df.csv')
beltran_2017.columns.values[1] = 'DATE'
beltran_2017.columns.values[2] = '2017 Batting Averages'
beltran_2017.drop('Unnamed: 0', axis = 1)
```

```python
frames = [dates, beltran_2012, beltran_2013, beltran_2014, beltran_2015, beltran_2016_1, beltran_2016_2, beltran_2017]
```

```python
df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['DATE'], how='outer'), frames)
```

```python
drops = ['Unnamed: 0_x', 'Unnamed: 0_y', 'Unnamed: 0']
```

```python
beltran_batting = df_merged.drop(drops, axis = 1)
```

```python
beltran_batting.to_csv('data_aggregated/beltran_batting.csv', index = False)
```
