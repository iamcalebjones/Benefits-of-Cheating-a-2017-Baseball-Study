```python
import pandas as pd
from functools import reduce

pd.set_option('display.max_rows', None)

dates = pd.read_csv('csv_from_scrape/baseball_dates_index.csv')
```

```python
correa_2015 = pd.read_csv('csv_from_scrape/correa_2015_df.csv')
correa_2015.columns.values[1] = 'DATE'
correa_2015.columns.values[2] = '2015 Batting Averages'
correa_2015.drop('Unnamed: 0', axis = 1)
```

```python
correa_2016 = pd.read_csv('csv_from_scrape/correa_2016_df.csv')
correa_2016.columns.values[1] = 'DATE'
correa_2016.columns.values[2] = '2016 Batting Averages'
correa_2016.drop('Unnamed: 0', axis = 1)
```

```python
correa_2017 = pd.read_csv('csv_from_scrape/correa_2017_df.csv')
correa_2017.columns.values[1] = 'DATE'
correa_2017.columns.values[2] = '2017 Batting Averages'
correa_2017.drop('Unnamed: 0', axis = 1)
```

```python
frames = [dates, correa_2015, correa_2016, correa_2017]
drops = ['Unnamed: 0_x', 'Unnamed: 0_y', 'Unnamed: 0']
```

```python
df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['DATE'], how='outer'), frames)
```

```python
correa_batting = df_merged.drop(drops, axis = 1)
```

```python
correa_batting.to_csv('data_aggregated/correa_batting.csv', index=False)
```
