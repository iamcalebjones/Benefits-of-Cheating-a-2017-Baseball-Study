```python
import pandas as pd
from functools import reduce
```

```python
pd.set_option('display.max_rows', None)
```

```python
dates = pd.read_csv('csv_from_scrape/baseball_dates_index.csv')
```

```python
bregman_2016 = pd.read_csv('csv_from_scrape/bregman_2016_df.csv')
bregman_2016.columns.values[1] = 'DATE'
bregman_2016.columns.values[2] = '2016 Batting Averages'
bregman_2016.drop('Unnamed: 0', axis = 1)
```

```python
bregman_2017 = pd.read_csv('csv_from_scrape/bregman_2017_df.csv')
bregman_2017.columns.values[1] = 'DATE'
bregman_2017.columns.values[2] = '2017 Batting Averages'
bregman_2017.drop('Unnamed: 0', axis = 1)
```

```python
frames = [dates, bregman_2016, bregman_2017]
```

```python
df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['DATE'], how='outer'), frames)
```

```python
drops = ['Unnamed: 0_x', 'Unnamed: 0_y']
```

```python
bregman_batting = df_merged.drop(drops, axis = 1)
```

```python
bregman_batting.to_csv('data_aggregated/bregman_batting.csv', index=False)
```
