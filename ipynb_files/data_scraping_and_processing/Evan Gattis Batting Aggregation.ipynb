```python
import pandas as pd
from functools import reduce

pd.set_option('display.max_rows', None)

dates = pd.read_csv('csv_from_scrape/baseball_dates_index.csv')
```

```python
gattis_2013 = pd.read_csv('csv_from_scrape/gattis_2013_df.csv')
gattis_2013.columns.values[1] = 'DATE'
gattis_2013.columns.values[2] = '2013 Batting Averages'
gattis_2013.drop('Unnamed: 0', axis = 1)
```

```python
gattis_2014 = pd.read_csv('csv_from_scrape/gattis_2014_df.csv')
gattis_2014.columns.values[1] = 'DATE'
gattis_2014.columns.values[2] = '2014 Batting Averages'
gattis_2014.drop('Unnamed: 0', axis = 1)
```

```python
gattis_2015 = pd.read_csv('csv_from_scrape/gattis_2015_df.csv')
gattis_2015.columns.values[1] = 'DATE'
gattis_2015.columns.values[2] = '2015 Batting Averages'
gattis_2015.drop('Unnamed: 0', axis = 1)
```

```python
gattis_2016 = pd.read_csv('csv_from_scrape/gattis_2016_df.csv')
gattis_2016.columns.values[1] = 'DATE'
gattis_2016.columns.values[2] = '2016 Batting Averages'
gattis_2016.drop('Unnamed: 0', axis = 1)
```

```python
gattis_2017 = pd.read_csv('csv_from_scrape/gattis_2017_df.csv')
gattis_2017.columns.values[1] = 'DATE'
gattis_2017.columns.values[2] = '2017 Batting Averages'
gattis_2017.drop('Unnamed: 0', axis = 1)
```

```python
frames = [dates, gattis_2013, gattis_2014, gattis_2015, gattis_2016, gattis_2017]
drops = ['Unnamed: 0_x', 'Unnamed: 0_y', 'Unnamed: 0']
```

```python
df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['DATE'], how='outer'), frames)
```

```python
gattis_batting = df_merged.drop(drops, axis = 1)
```

```python
gattis_batting.to_csv('data_aggregated/gattis_batting.csv', index=False)
```
