```python
import pandas as pd
import requests
import pprint
from pymongo import MongoClient
from bs4 import BeautifulSoup
import json
import time
from lxml import html
from io import StringIO
from datetime import datetime
```

# Evan Gattis

Gattis played for the Atlanta Braves from 2013-2014, and the Houston Astros from 2015-2017

## 2013: Braves

```python
# batting averages

xpath = "//pre[5]/text()"
url = 'https://www.retrosheet.org/boxesetc/2013/Igatte0010012013.htm'
r = requests.get(url)
page = html.parse(StringIO(r.text))
gattis_2013 = page.xpath(xpath)

gattis_2013_game_data_only = []
for idx, row in enumerate(gattis_2013):
    if idx % 2 == 0:
        gattis_2013_game_data_only.append(row)
        

gattis_2013_game_data_cleaned = []
for row in gattis_2013_game_data_only[1:]:
    if row[112:119] == '   Date':
        gattis_2013_game_data_cleaned.append(row[0:112])
    else:
        gattis_2013_game_data_cleaned.append(row)
        
gattis_2013_averages = [float(row[87:91]) for row in gattis_2013_game_data_cleaned]


# dates played this year

soup = BeautifulSoup(r.text, 'html.parser')

link_text = []
for a in soup.find_all('a'):
    link_text.append(a.string)

gattis_2013_dates = []
for i in link_text:
    if '2013' in i:
        gattis_2013_dates.append(i)

gattis_2013_dates_cleaned = []
for date in gattis_2013_dates:
    gattis_2013_dates_cleaned.append(date.replace(' ', '0'))
    
gattis_2013_dates_final = []
for date in gattis_2013_dates_cleaned:
    gattis_2013_dates_final.append(datetime.strftime(datetime.strptime(date, '%m-%d-%Y'), '%m-%d'))
    
gattis_2013_df = pd.DataFrame(gattis_2013_averages, gattis_2013_dates_final).reset_index()
gattis_2013_df.columns = ['2013', 'Batting Average']
gattis_2013_df
```

```python
gattis_2013_df.plot('2013', 'Batting Average')
```
