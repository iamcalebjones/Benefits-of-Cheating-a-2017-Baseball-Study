```python
import pandas as pd
from functools import reduce

pd.set_option('display.max_rows', None)

dates = pd.read_csv('csv_from_scrape/baseball_dates_index.csv')
```

```python
marisnick_2013 = pd.read_csv('csv_from_scrape/marisnick_2013_df.csv')
marisnick_2013.columns.values[1] = 'DATE'
marisnick_2013.columns.values[2] = '2013 Batting Averages'
marisnick_2013.drop('Unnamed: 0', axis = 1)
```

```python
marisnick_2014_1 = pd.read_csv('csv_from_scrape/marisnick_2014_1_df.csv')
marisnick_2014_1.columns.values[1] = 'DATE'
marisnick_2014_1.columns.values[2] = '2014 Batting Averages 1'
marisnick_2014_1.drop('Unnamed: 0', axis = 1)
```

```python
marisnick_2014_2 = pd.read_csv('csv_from_scrape/marisnick_2014_2_df.csv')
marisnick_2014_2.columns.values[1] = 'DATE'
marisnick_2014_2.columns.values[2] = '2014 Batting Averages 2'
marisnick_2014_2.drop('Unnamed: 0', axis = 1)
```

```python
marisnick_2015 = pd.read_csv('csv_from_scrape/marisnick_2015_df.csv')
marisnick_2015.columns.values[1] = 'DATE'
marisnick_2015.columns.values[2] = '2015 Batting Averages'
marisnick_2015.drop('Unnamed: 0', axis = 1)
```

```python
marisnick_2016 = pd.read_csv('csv_from_scrape/marisnick_2016_df.csv')
marisnick_2016.columns.values[1] = 'DATE'
marisnick_2016.columns.values[2] = '2016 Batting Averages'
marisnick_2016.drop('Unnamed: 0', axis = 1)
```

```python
marisnick_2017 = pd.read_csv('csv_from_scrape/marisnick_2017_df.csv')
marisnick_2017.columns.values[1] = 'DATE'
marisnick_2017.columns.values[2] = '2017 Batting Averages'
marisnick_2017.drop('Unnamed: 0', axis = 1)
```

```python
frames = [dates, marisnick_2013, marisnick_2014_1, marisnick_2014_2, marisnick_2015, marisnick_2016, marisnick_2017]
drops = ['Unnamed: 0_x', 'Unnamed: 0_y']
```

```python
df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['DATE'], how='outer'), frames)
```

```python
marisnick_batting = df_merged.drop(drops, axis = 1)
```

```python
marisnick_batting.to_csv('data_aggregated/marisnick_batting.csv', index=False)
```
