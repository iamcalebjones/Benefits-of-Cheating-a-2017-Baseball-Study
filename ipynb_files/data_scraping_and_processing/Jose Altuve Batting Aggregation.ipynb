```python
import pandas as pd
from functools import reduce
```

```python
pd.set_option('display.max_rows', None)
```

```python
dates = pd.read_csv('../csv_from_scrape/baseball_dates_index.csv')
```

```python
dates
frames = [dates]
```

```python
altuve_2011 = pd.read_csv('../csv_from_scrape/altuve_2011_df.csv')
altuve_2011.columns.values[1] = 'DATE'
altuve_2011.columns.values[2] = '2011 Batting Averages'
altuve_2011.drop('Unnamed: 0', axis = 1)
frames.append(altuve_2011)
```

```python
altuve_2012 = pd.read_csv('../csv_from_scrape/altuve_2012_df.csv')
altuve_2012.columns.values[1] = 'DATE'
altuve_2012.columns.values[2] = '2012 Batting Averages'
altuve_2012.drop('Unnamed: 0', axis = 1)
frames.append(altuve_2012)
```

```python
altuve_2013 = pd.read_csv('../csv_from_scrape/altuve_2013_df.csv')
altuve_2013.columns.values[1] = 'DATE'
altuve_2013.columns.values[2] = '2013 Batting Averages'
altuve_2013.drop('Unnamed: 0', axis = 1)
frames.append(altuve_2013)
```

```python
altuve_2014 = pd.read_csv('../csv_from_scrape/altuve_2014_df.csv')
altuve_2014.columns.values[1] = 'DATE'
altuve_2014.columns.values[2] = '2014 Batting Averages'
altuve_2014.drop('Unnamed: 0', axis = 1)
frames.append(altuve_2014)
```

```python
altuve_2015 = pd.read_csv('../csv_from_scrape/altuve_2015_df.csv')
altuve_2015.columns.values[1] = 'DATE'
altuve_2015.columns.values[2] = '2015 Batting Averages'
altuve_2015.drop('Unnamed: 0', axis = 1)
frames.append(altuve_2015)
```

```python
altuve_2016 = pd.read_csv('../csv_from_scrape/altuve_2016_df.csv')
altuve_2016.columns.values[1] = 'DATE'
altuve_2016.columns.values[2] = '2016 Batting Averages'
altuve_2016.drop('Unnamed: 0', axis = 1)
frames.append(altuve_2016)
```

```python
altuve_2017 = pd.read_csv('../csv_from_scrape/altuve_2017_df.csv')
altuve_2017.columns.values[1] = 'DATE'
altuve_2017.columns.values[2] = '2017 Batting Averages'
altuve_2017.drop('Unnamed: 0', axis = 1)
frames.append(altuve_2017)
```

```python
df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['DATE'], how='outer'), frames)
```

```python
df_merged
```

```python
drops = ['Unnamed: 0_x', 'Unnamed: 0_y', 'Unnamed: 0']
```

```python
altuve_batting = df_merged.drop(drops, axis = 1)
```

```python
altuve_batting
```

```python
altuve_batting.to_csv('../data_aggregated/altuve_batting.csv', index=False)
```
