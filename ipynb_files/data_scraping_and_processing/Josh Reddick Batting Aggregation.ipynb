```python
frames = [dates, reddick_2009, reddick_2010, reddick_2011, reddick_2012, reddick_2013, reddick_2014, reddick_2015, 
          reddick_2016_1, reddick_2016_2, reddick_2017]
drops = ['Unnamed: 0_x', 'Unnamed: 0_y']
```

```python
df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['DATE'], how='outer'), frames)
```

```python
reddick_batting = df_merged.drop(drops, axis = 1)
```

```python
reddick_batting.to_csv('data_aggregated/reddick_batting.csv', index=False)
```
