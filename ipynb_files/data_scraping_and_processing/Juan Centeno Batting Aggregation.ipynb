```python
import pandas as pd
from functools import reduce

pd.set_option('display.max_rows', None)

dates = pd.read_csv('csv_from_scrape/baseball_dates_index.csv')
```

```python
centeno_2013 = pd.read_csv('csv_from_scrape/centeno_2013_df.csv')
centeno_2013.columns.values[1] = 'DATE'
centeno_2013.columns.values[2] = '2013 Batting Averages'
centeno_2013.drop('Unnamed: 0', axis = 1)
```

```python
centeno_2014 = pd.read_csv('csv_from_scrape/centeno_2014_df.csv')
centeno_2014.columns.values[1] = 'DATE'
centeno_2014.columns.values[2] = '2014 Batting Averages'
centeno_2014.drop('Unnamed: 0', axis = 1)
```

```python
centeno_2015 = pd.read_csv('csv_from_scrape/centeno_2015_df.csv')
centeno_2015.columns.values[1] = 'DATE'
centeno_2015.columns.values[2] = '2015 Batting Averages'
centeno_2015.drop('Unnamed: 0', axis = 1)
```

```python
centeno_2016 = pd.read_csv('csv_from_scrape/centeno_2016_df.csv')
centeno_2016.columns.values[1] = 'DATE'
centeno_2016.columns.values[2] = '2016 Batting Averages'
centeno_2016.drop('Unnamed: 0', axis = 1)
```

```python
centeno_2017 = pd.read_csv('csv_from_scrape/centeno_2017_df.csv')
centeno_2017.columns.values[1] = 'DATE'
centeno_2017.columns.values[2] = '2017 Batting Averages'
centeno_2017.drop('Unnamed: 0', axis = 1)
```

```python
frames = [dates, centeno_2013, centeno_2014, centeno_2015, centeno_2016, centeno_2017]
drops = ['Unnamed: 0_x', 'Unnamed: 0_y', 'Unnamed: 0']
```

```python
df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['DATE'], how='outer'), frames)
```

```python
centeno_batting = df_merged.drop(drops, axis = 1)
```

```python
centeno_batting.to_csv('data_aggregated/centeno_batting.csv', index=False)
```
