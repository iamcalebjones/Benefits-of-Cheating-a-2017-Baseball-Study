```python
import pandas as pd
import requests
import pprint
from pymongo import MongoClient
from bs4 import BeautifulSoup
import json
import time
from lxml import html
from io import StringIO
from datetime import datetime
```

# Max Stassi

```python
urls = ['https://www.retrosheet.org/boxesetc/2013/Istasm0010012013.htm',
        'https://www.retrosheet.org/boxesetc/2014/Istasm0010022014.htm',
        'https://www.retrosheet.org/boxesetc/2015/Istasm0010032015.htm',
        'https://www.retrosheet.org/boxesetc/2017/Istasm0010052017.htm']

csv_names = ['csv_from_scrape/stassi_2013_df.csv',
             'csv_from_scrape/stassi_2014_df.csv',
             'csv_from_scrape/stassi_2015_df.csv',
             'csv_from_scrape/stassi_2017_df.csv']

years = ['2013', '2014', '2015', '2017']
```

```python
len(urls) == len(csv_names) == len(years)
```

```python
for idx, url in enumerate(urls):
    
    # batting averages

    xpath = "//pre[5]/text()"
    r = requests.get(url)
    page = html.parse(StringIO(r.text))
    batter = page.xpath(xpath)
    
    batter_game_data_only = []
    for a, row in enumerate(batter):
        if a % 2 == 0:
            batter_game_data_only.append(row)
    
    batter_game_data_cleaned = []
    for row in batter_game_data_only[1:]:
        if len(row) > len(batter_game_data_only[2]):
            batter_game_data_cleaned.append(row[0:112])
        else:
            batter_game_data_cleaned.append(row)
        
    batter_averages = [float(row[87:91]) for row in batter_game_data_cleaned]

    # dates played this year

    soup = BeautifulSoup(r.text, 'html.parser')
    
    link_text = []
    for a in soup.find_all('a'):
        link_text.append(a.string)
        
    batter_dates = []
    for i in link_text:
        if years[idx] in i:
            batter_dates.append(i)
            
    batter_dates_cleaned = []
    for date in batter_dates:
        batter_dates_cleaned.append(date.replace(' ', '0'))
        
    batter_dates_final = []
    for date in batter_dates_cleaned:
        batter_dates_final.append(datetime.strftime(datetime.strptime(date, '%m-%d-%Y'), '%m-%d'))
        
    batter_df = pd.DataFrame(batter_averages, batter_dates_final).reset_index()
    batter_df.columns = [years[idx], 'Batting Average']
    batter_df.to_csv(csv_names[idx])
    batter_df.plot(years[idx], 'Batting Average')
    
    print(years[idx] + ' complete')
```

```python
xpath = "//pre[5]/text()"
r = requests.get('https://www.retrosheet.org/boxesetc/2016/Istasm0010042016.htm')
page = html.parse(StringIO(r.text))
batter = page.xpath(xpath)

batter_game_data_only = []
for a, row in enumerate(batter):
    if a % 2 == 0:
        batter_game_data_only.append(row)

batter_game_data_cleaned = []
for row in batter_game_data_only[2:]:
    batter_game_data_cleaned.append(row[0:100])

batter_averages = [float(row[87:91]) for row in batter_game_data_cleaned]

# dates played this year

soup = BeautifulSoup(r.text, 'html.parser')

link_text = []
for a in soup.find_all('a'):
    link_text.append(a.string)

batter_dates = []
for i in link_text:
    if '2016' in i:
        batter_dates.append(i)

batter_dates_cleaned = []
for date in batter_dates[1:]:
    batter_dates_cleaned.append(date.replace(' ', '0'))

batter_dates_final = []
for date in batter_dates_cleaned:
    batter_dates_final.append(datetime.strftime(datetime.strptime(date, '%m-%d-%Y'), '%m-%d'))

batter_df = pd.DataFrame(batter_averages, batter_dates_final).reset_index()
batter_df.columns = ['2016', 'Batting Average']
batter_df.to_csv('csv_from_scrape/stassi_2016_df.csv')
batter_df.plot('2016', 'Batting Average')
```
